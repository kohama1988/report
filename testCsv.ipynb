```python
import re
import pandas as pd
import re
from pptx import Presentation
import neologdn
import unicodedata
import numpy as np
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
```

```python
text = 'I want to \n have \n\n\n some food'
re.sub('\s+', ' ', text)
```

```python
def extractExcel(filepath):
    res = ''
    dfp = pd.read_excel(filepath, na_filter=False, sheet_name=None)
    for i in dfp:
        dfm = pd.read_excel(filepath, na_filter=False, sheet_name=i)
        text = re.sub('\s+', ' ', dfm.to_string(header=None, index=None, na_rep='', index_names=False))
        res += text
    return res 
```

```python
def cleanText(string):
    # remove unicode and crlf
    text = re.sub('\s+', '', re.sub('\n|,','', string)).encode('cp932', errors='ignore').decode('cp932')
    text = text.replace('\u3000','')
    # string normalize
    text = neologdn.normalize(text)
    text = unicodedata.normalize('NFKC', text)
    return text
```

```python
res = extractExcel('T-8028-1 樹脂インレットチューブ SPEC 評価結果.xlsx')
res = cleanText(res)
res[:100]
```
