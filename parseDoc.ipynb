```python
import re
from pptx import Presentation
import neologdn
import unicodedata
import numpy as np
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
```

## pptx

```python
# Open the PowerPoint file
prs = Presentation('T-8003_2 水系冷却配管提案(20220926).pptx')
res = ''
# Iterate through all slides in the file
for slide in prs.slides:
    for shape in slide.shapes:
        if not shape.has_text_frame:
            continue
        # Iterate through all the paragraphs in the text frame
        for paragraph in shape.text_frame.paragraphs:
            # Iterate through all the runs in the paragraph
            for run in paragraph.runs:
                # Print the text of the run
                res += run.text
```

```python
# pptxのすべての文字を抽出する関数
def extractPPTX(file):
    prs = Presentation(file)
    res = ''
    for slide in prs.slides:
        for shape in slide.shapes:
            if not shape.has_text_frame:
                continue
            for paragraph in shape.text_frame.paragraphs:
                for run in paragraph.runs:
                    res += run.text
    return res
```

```python
def cleanText(string):
    # remove unicode and crlf
    text = re.sub('\s+', '', re.sub('\n','', string)).encode('cp932', errors='ignore').decode('cp932')
    text = text.replace('\u3000','')
    # string normalize
    text = neologdn.normalize(text)
    text = unicodedata.normalize('NFKC', text)
    return text
```

```python
text = extractPPTX('T-8003_2 水系冷却配管提案(20220926).pptx')
cleanText(text)
```

## ppt

```python
import win32com
```

```python
def extractPPT(file):
    res = ''
    app = win32com.client.Dispatch("PowerPoint.Application")
    pptsel = app.Presentations.Open(file, WithWindow=1)
    for slide in pptsel.Slides:
        for shape in slide.Shapes:
            if shape.HasTextFrame:
                if shape.TextFrame.HasText:
                    res += shape.TextFrame.TextRange.Text
    pptsel.Close()
    app.Quit()
    return res
```

```python
text = extractPPT('T-8003_2 水系冷却配管提案(20220926).ppt')
cleanText(text)
```

## Excel

```python
import pandas as pd
```

```python
def extractExcel(filepath):
    res = ''
    dfp = pd.read_excel(filepath, na_filter=False, sheet_name=None)
    for i in dfp:
        dfm = pd.read_excel(filepath, na_filter=False, sheet_name=i)
        text = re.sub('\s+', ' ', dfm.to_string(header=None, index=None, na_rep='', index_names=False), re.M)
        res += text
    return res 
```

```python
extractExcel('T-8004 樹脂インレットチューブ SPEC と仕様に関する質問事項 .xls')
```

```python
import os
```

```python
path = r'\\kfs04\share3\4182-jihogi3\開発データ\01_製品設計\015_報告書\0151_ユーザー提出報告書\トヨタ\T-8001～8500'
def getContent(path):
    res = []
    os.chdir(path)
    for file in os.listdir(path):
        try:
            if file.endswith('.pptx'):
                print('pptx: ', file)
                text = extractPPTX(file)
                text = cleanText(text)
                res.append([cleanText(file), text])
            elif file.endswith('.ppt'):
                print('ppt: ', file)
                text = extractPPT(file)
                text = cleanText(text)
                res.append([cleanText(file), text])
            elif file.endswith('.xlsx') or file.endswith('.xls'):
                print('excel: ', file)
                text = extractExcel(file)
                text = cleanText(text)
                res.append([cleanText(file), text])
        except Exception as e:
            print(e)
            print('*'*50)
    return res
res = getContent(path)
```
